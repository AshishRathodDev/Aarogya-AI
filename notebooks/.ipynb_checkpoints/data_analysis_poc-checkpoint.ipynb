```python
print("\n--- Statistical Summary of Test Results ('result' column) ---")
display(df['result'].describe())
```

```python
print("\n--- Top 15 Most Common Tests Found ---")
# .value_counts() humein batata hai ki kaun sa test kitni baar aaya
display(df['test_name'].value_counts().head(15))
```

```python
print("\n--- Visualizing Outliers in Top 10 Tests ---")
```

```python
top_10_tests = df['test_name'].value_counts().head(10).index
```

```python
df_top_10 = df[df['test_name'].isin(top_10_tests)]
```

```python
# Plot banao
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_top_10, x='test_name', y='result')
plt.xticks(rotation=45)
plt.title('Box Plot of Results for Top 10 Tests')
plt.xlabel('Test Name')
plt.ylabel('Result Value')
plt.tight_layout()
plt.show()
```
